### STRING

### 1. What is a String?

* A string is a sequence of characters enclosed in single quotes ('...'), double quotes ("..."), or triple quotes ('''...''' for multi-line strings).

EXAMPLE - 

name = "Dilawezdaniya"
college = 'PIMR'

### 2. String Slicing

* Slicing allows you to get a part of the string using the index. Remember, Python indexing starts at 0.
Syntax: string[start : end : step]

* Example:

s = "Python"

print(s[0:2])  # Output: 'Py' (Index 0 and 1, stops before 2)

print(s[::-1]) # Output: 'nohtyP' (Reverses the string!)

### 3. Escape Characters

* As you noted in checklist , escape characters help you format text:

* \n (Newline): Moves the text following it to a new line.

* \t (Tab): Inserts a horizontal tab space, which is useful for aligning text.

* \\ (Backslash): Allows you to insert a literal backslash into the string.

* \' and \" (Quotes): These allow you to use quotes inside a string that is already defined by the same type of quote.

Example: text = "He said, \"Hello!\""

### 4. Useful String Methods

* These are built-in functions :


* upper() / .lower(): Changes the case of the string.

* .strip(): Removes extra spaces from the beginning and end.

* replace("old", "new"): Replaces a part of the string.

* find("char"): Returns the index of the first time that character appears.

* len(string): Gives the total length of the string.

### 5 String Formatting (f-strings)
 
*  F-strings are the best way to include variables inside text.
EXAMPLE - 

topic = "Strings"
print(f"Today I am learning {topic} in Python!")

* PROBLEM 2 - WRITE A PROGRAM TO FILL IN A LETTER TEMPLATE GIVEN BELOW WITH NAME AND DATA


```python
LETTER = '''
        DEAR <|NAME|>,
        YOU ARE SELECTED FOR THE POSITION OF PYTHON DEVELOPER.
        <|DATE|>
          '''
print(LETTER .replace("<|NAME|>", "DILAAWEZ").replace("<|DATE|>", "2027-06-01"))       
```
