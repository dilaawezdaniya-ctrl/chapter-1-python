### STRING

### 1. What is a String?

* A string is a sequence of characters enclosed in single quotes ('...'), double quotes ("..."), or triple quotes ('''...''' for multi-line strings).

EXAMPLE - 

name = "Dilawezdaniya"
college = 'PIMR'

### 2. String Slicing

* Slicing allows you to get a part of the string using the index. Remember, Python indexing starts at 0.
Syntax: string[start : end : step]

* Example:

s = "Python"
print(s[0:2])  # Output: 'Py' (Index 0 and 1, stops before 2)
print(s[::-1]) # Output: 'nohtyP' (Reverses the string!)

### 3. Escape Characters

* As you noted in your checklist yesterday, escape characters help you format text:

* \n: New line (make sure to use the backslash \ as you correctly identified in your summary!).

* \t: Tab space.

### 4. Useful String Methods

* These are built-in functions you should practice in your strings.ipynb file:

* upper() / .lower(): Changes the case of the string.

* .strip(): Removes extra spaces from the beginning and end.

* replace("old", "new"): Replaces a part of the string.

* find("char"): Returns the index of the first time that character appears.

* len(string): Gives the total length of the string.
