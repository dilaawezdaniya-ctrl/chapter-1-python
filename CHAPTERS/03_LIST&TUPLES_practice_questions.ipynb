### Q8. Student Report Card
```python
students = [("Rahul", 88), ("Priya", 95), ("Aman", 76), ("Sneha", 91)]
```
1. Print each student's name and marks
2. Find the student with the highest marks
3. Sort by marks in descending order

````python
# ✏️ Your Solution — Q8
students = [("Rahul", 88), ("Priya", 95), ("Aman", 76), ("Sneha", 91)]

# 1. Print each student

# 2. Find topper

# 3. Sort by marks descending
````

````python
# 💡 Hint
print("→ for name, marks in students:  ← unpack tuple in loop")
print("→ max(students, key=lambda s: s[1])")
print("→ sorted(students, key=lambda s: s[1], reverse=True)")
````

````python
# ✅ Solution
students = [("Rahul", 88), ("Priya", 95), ("Aman", 76), ("Sneha", 91)]

for name, marks in students:
    print(f"{name}: {marks}")

topper = max(students, key=lambda s: s[1])
print(f"\nTopper: {topper[0]} with {topper[1]} marks")

ranked = sorted(students, key=lambda s: s[1], reverse=True)
print("\nRankings:")
for i, (name, marks) in enumerate(ranked, 1):
    print(f"  Rank {i}: {name} — {marks}")
````

### Q10. List Comprehension Filter
```python
numbers = [3, 8, 15, 4, 7, 12, 1, 20, 6, 11]
```
Using **list comprehension**:
1. Even numbers only
2. Squares of odd numbers
3. Numbers greater than 10, multiplied by 2

````python
# ✏️ Your Solution — Q10
numbers = [3, 8, 15, 4, 7, 12, 1, 20, 6, 11]

# 1. Evens

# 2. Squares of odds

# 3. Numbers > 10, doubled
````

````python
# 💡 Hint
print("→ Syntax: [expression for item in list if condition]")
print("→ Even:  num % 2 == 0")
print("→ Odd:   num % 2 != 0")
````

````python
# ✅ Solution
numbers = [3, 8, 15, 4, 7, 12, 1, 20, 6, 11]

evens       = [n for n in numbers if n % 2 == 0]
odd_squares = [n ** 2 for n in numbers if n % 2 != 0]
big_doubled = [n * 2 for n in numbers if n > 10]

print("Evens:",       evens)
print("Odd squares:", odd_squares)
print("Big doubled:", big_doubled)
````

### Q13. Remove Duplicates (preserving order)
```python
tags = ["python", "code", "python", "list", "code", "tuple", "python"]
# Expected: ["python", "code", "list", "tuple"]
```
Remove duplicates while keeping the **original order**.

````python
# ✏️ Your Solution — Q13
tags = ["python", "code", "python", "list", "code", "tuple", "python"]
````

````python
# 💡 Hint
print("→ Keep a 'seen' set")
print("→ Only add item to result if not already in seen")
print("→ A plain set() removes duplicates but loses order")
````

````python
# ✅ Solution
tags = ["python", "code", "python", "list", "code", "tuple", "python"]

seen, unique_tags = set(), []
for tag in tags:
    if tag not in seen:
        unique_tags.append(tag)
        seen.add(tag)

print(unique_tags)   # ['python', 'code', 'list', 'tuple']

# One-liner
print(list(dict.fromkeys(tags)))
````
