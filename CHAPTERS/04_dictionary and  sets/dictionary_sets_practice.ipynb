### Q3. Loop a Dictionary
Given:
```python
marks = {"Maths": 95, "Science": 88, "English": 91, "Hindi": 78}
```
1. Print all subjects
2. Print all marks
3. Print each subject and its marks in format: `Maths → 95`
4. Find the subject with the highest marks

````python
# ✏️ Your Solution — Q3
marks = {"Maths": 95, "Science": 88, "English": 91, "Hindi": 78}

# 1. Print all subjects

# 2. Print all marks

# 3. Print subject → marks

# 4. Highest marks subject
````

````python
# 💡 Hint
print("→ dict.keys()              all keys")
print("→ dict.values()            all values")
print("→ for k, v in dict.items() key-value pairs")
print("→ max(dict, key=dict.get)  key with max value")
````

````python
# ✅ Solution
marks = {"Maths": 95, "Science": 88, "English": 91, "Hindi": 78}

print(list(marks.keys()))
print(list(marks.values()))

for subject, mark in marks.items():
    print(f"{subject} → {mark}")

best = max(marks, key=marks.get)
print(f"\nHighest: {best} ({marks[best]})")
````

Find:

4. Symmetric difference — items in either but not both

### Q8. Word Frequency Counter
Write a program that counts how many times each word appears in a sentence.
```python
sentence = "the cat sat on the mat the cat sat"
```
1. Count each word's frequency
2. Print word → count for each word
3. Find the most common word

````python
# ✏️ Your Solution — Q8
sentence = "the cat sat on the mat the cat sat"

# 1. Count frequencies

# 2. Print word → count

# 3. Most common word
````

````python
# 💡 Hint
print("→ sentence.split()              splits into word list")
print("→ dict.get(word, 0) + 1         count safely")
print("→ max(dict, key=dict.get)        key with highest value")
````

````python
# ✅ Solution
sentence = "the cat sat on the mat the cat sat"
words    = sentence.split()

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

for word, count in freq.items():
    print(f"{word} → {count}")

most_common = max(freq, key=freq.get)
print(f"\nMost common: '{most_common}' ({freq[most_common]} times)")
````

````python
# 💡 Hint
print("→ db[roll] = {'name':..., 'marks':[...]}")
print("→ sum(marks)/len(marks)  for average")
print("→ max() with key= for highest")
````

Using dictionary comprehension:

```python

```


5. Find the most expensive item

### Q13. Group Students by Grade
Given a list of `(name, marks)` tuples, group students into grade buckets using a dictionary:
- `A` → marks >= 90
- `B` → marks >= 75
- `C` → marks below 75

```python
results = [("Priya",95),("Rahul",82),("Aman",68),("Sneha",91),("Vikram",74),("Neha",88)]
```

````python
# ✏️ Your Solution — Q13
results = [("Priya",95),("Rahul",82),("Aman",68),("Sneha",91),("Vikram",74),("Neha",88)]

# Group into A, B, C grades
````

````python
# 💡 Hint
print("→ Start with grades = {'A': [], 'B': [], 'C': []}")
print("→ Loop through results, check marks, append name to right grade")
````

````python
# ✅ Solution
results = [("Priya",95),("Rahul",82),("Aman",68),("Sneha",91),("Vikram",74),("Neha",88)]

grades = {"A": [], "B": [], "C": []}

for name, marks in results:
    if marks >= 90:
        grades["A"].append(name)
    elif marks >= 75:
        grades["B"].append(name)
    else:
        grades["C"].append(name)

for grade, students in grades.items():
    print(f"Grade {grade}: {students}")

# Grade A: ['Priya', 'Sneha']
# Grade B: ['Rahul', 'Neha']
# Grade C: ['Aman', 'Vikram']
````
