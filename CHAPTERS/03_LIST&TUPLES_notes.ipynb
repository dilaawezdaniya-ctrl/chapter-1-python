## 1. What are they? <a id='1'></a>

Both **lists** and **tuples** are:
- Ordered collections — items have a fixed position/index
- Able to store any mix of data types (`int`, `str`, `float`, etc.)
- Iterable — you can loop over them
- Built into Python — no imports needed

The **one big difference** → mutability.

---
## 2. The Core Difference — Mutability <a id='2'></a>

| | List | Tuple |
|---|---|---|
| Can be changed after creation? | ✅ Yes (mutable) | ❌ No (immutable) |
| Analogy | Whiteboard — erase & rewrite | Stone carving — permanent |

````python
# ✅ List — mutable
my_list = [1, 2, 3]
my_list[0] = 99
print(my_list)        # [99, 2, 3]

# ❌ Tuple — immutable
my_tuple = (1, 2, 3)
# my_tuple[0] = 99    # uncomment to see TypeError
````

---
## 3. Syntax <a id='3'></a>

```
List  →  square brackets  [ ]
Tuple →  parentheses      ( )
```

> ⚠️ **Single-element tuple** — the trailing comma is REQUIRED!

````python
fruits = ["apple", "banana", "cherry"]   # list
point  = (10, 20, 30)                    # tuple

# Single element
single_list  = [42]     # list ✅
single_tuple = (42,)    # tuple ✅  ← comma is required!
not_a_tuple  = (42)     # just an int ❌

print(type(single_list))   # <class 'list'>
print(type(single_tuple))  # <class 'tuple'>
print(type(not_a_tuple))   # <class 'int'>
````

## 4. Shared Features <a id='4'></a>
Both lists and tuples support: indexing, slicing, looping, `len()`, and membership checks.

````python
colors = ["red", "green", "blue"]   # list
point  = (10, 20, 30)               # tuple

# Indexing
print(colors[0])     # red
print(point[0])      # 10

# Negative indexing
print(colors[-1])    # blue
print(point[-1])     # 30

# Slicing
print(colors[1:])    # ['green', 'blue']
print(point[:2])     # (10, 20)

# Length
print(len(colors))   # 3
print(len(point))    # 3

# Membership
print("red" in colors)   # True
print(20 in point)       # True

# Count & Index
nums = [1, 2, 2, 3, 2]
print(nums.count(2))     # 3
print(nums.index(3))     # 3
````

## 5. List — Deep Dive <a id='5'></a>
Lists are **mutable** — you can add, remove, update, and reorder items freely.

````python
fruits = ["apple", "banana", "cherry"]

# ➕ Add items
fruits.append("mango")            # add to end
fruits.insert(1, "kiwi")          # add at index
fruits.extend(["grape", "lime"])  # add multiple
print("After adding:", fruits)

# ➖ Remove items
fruits.remove("banana")           # remove by value
popped = fruits.pop()             # remove & return last
print("After removing:", fruits)
print("Popped:", popped)

# 🔃 Reorder
nums = [3, 1, 4, 1, 5, 9]
nums.sort()                       # ascending
print("Sorted:", nums)
nums.sort(reverse=True)           # descending
print("Descending:", nums)
nums.reverse()                    # reverse in place
print("Reversed:", nums)
````

````python
# 🐍 List Comprehension — Pythonic way to build lists

# Traditional way
squares = []
for x in range(1, 6):
    squares.append(x ** 2)
print("Traditional:", squares)

# List comprehension — same result, one line
squares = [x ** 2 for x in range(1, 6)]
print("Comprehension:", squares)   # [1, 4, 9, 16, 25]

# With condition
evens = [x for x in range(20) if x % 2 == 0]
print("Evens:", evens)
````

## 6. Tuple — Deep Dive <a id='6'></a>
Tuples are **immutable** — once created, they cannot be modified.

````python
# Packing — parentheses are optional!
person = "Bob", 30, "Engineer"
print(type(person))   # <class 'tuple'>

# Unpacking
point = (10, 20, 30)
x, y, z = point
print(x, y, z)   # 10 20 30

# Swap without temp variable
a, b = 5, 10
a, b = b, a
print("Swapped:", a, b)   # 10 5

# Extended unpacking with *
first, *rest = (1, 2, 3, 4, 5)
print("First:", first)    # 1
print("Rest:", rest)      # [2, 3, 4, 5]

*start, last = (1, 2, 3, 4, 5)
print("Start:", start)    # [1, 2, 3, 4]
print("Last:", last)      # 5
````

````python
# Tuples as dictionary keys — lists CANNOT be dict keys!
seat_map = {
    (1, "A"): "Alice",
    (1, "B"): "Bob",
    (2, "C"): "Charlie",
}
print(seat_map[(1, "A")])   # Alice

# bad = {[1, "A"]: "Alice"}  # ❌ TypeError — uncomment to see
````

## 7. Real-World Examples <a id='7'></a>

````python
# 🛒 Shopping Cart → List (changes constantly)
cart = ["milk", "eggs"]
cart.append("bread")
cart.remove("eggs")
print("Cart:", cart)
````

````python
# ✅ Todo List → List
tasks = ["reply email", "team meeting", "submit report"]
tasks.append("deploy to production")
tasks.pop(0)   # first task done
print("Tasks:", tasks)
````

````python
# 🗺️ GPS Coordinates → Tuple (a city never moves)
delhi  = (28.6139, 77.2090)
mumbai = (19.0760, 72.8777)
lat, lon = delhi
print(f"Delhi → Latitude: {lat}, Longitude: {lon}")
````

````python
# 🎨 RGB Colour Constants → Tuple
RED   = (255, 0,   0)
GREEN = (0,   255, 0)
BLUE  = (0,   0,   255)
r, g, b = RED
print(f"Red → R={r}, G={g}, B={b}")
````

````python
# 🗄️ Database Record → Tuple (read-only)
student_row = (101, "Priya Sharma", "Computer Science", 9.1)
student_id, name, dept, cgpa = student_row
print(f"{name} | {dept} | CGPA: {cgpa}")
````

````python
# 🎬 Cinema Seat Booking → Tuple as dict key
bookings = {
    (3, "A"): "Rahul",
    (3, "B"): "Sneha",
    (5, "C"): "Vikram",
}
print(f"Seat (3, 'A') → {bookings[(3, 'A')]}")
````

````python
# 🔁 Function returning multiple values → Tuple
def get_stats(data):
    return min(data), max(data), sum(data) / len(data)

marks = [88, 76, 91, 85]
low, high, avg = get_stats(marks)
print(f"Min: {low} | Max: {high} | Avg: {avg}")
````

## 8. Performance Comparison <a id='8'></a>
Tuples are **faster** and use **less memory** than lists because Python knows they'll never change.

````python
import sys
import timeit

my_list  = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)

# Memory
print(f"List  size: {sys.getsizeof(my_list)} bytes")
print(f"Tuple size: {sys.getsizeof(my_tuple)} bytes")

# Speed
list_time  = timeit.timeit("[1, 2, 3, 4, 5]", number=5_000_000)
tuple_time = timeit.timeit("(1, 2, 3, 4, 5)", number=5_000_000)
print(f"\nList  creation time: {list_time:.3f}s")
print(f"Tuple creation time: {tuple_time:.3f}s")
print("Tuple is faster! ✅")
````

## 9. Quick Reference Table <a id='9'></a>

| Feature | List `[ ]` | Tuple `( )` |
|---|---|---|
| Mutable | ✅ Yes | ❌ No |
| Syntax | `[1, 2, 3]` | `(1, 2, 3)` |
| Indexing | ✅ | ✅ |
| Slicing | ✅ | ✅ |
| Looping | ✅ | ✅ |
| `append()` | ✅ | ❌ |
| `remove()` | ✅ | ❌ |
| `sort()` | ✅ | ❌ |
| As dict key | ❌ | ✅ |
| Memory | More | Less |
| Speed | Slower | Faster |
| Use when | Data changes | Data is fixed |

---

### 🏁 The Golden Rule

```
"Will this data need to change after I create it?"

    YES  →  List  [ ]
    NO   →  Tuple ( )
```
